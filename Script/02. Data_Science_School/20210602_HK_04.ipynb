```python
#2021.06.02. WED
#Hankyeong

##CH05-1. 데이터 시각화 기초
#00. 패키지 호출
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm 
import numpy as np 

#01. Matplotlib에서 한글 폰트 사용하기.
#(1) 폰트가 시스템에 존재하는지 확인하기.
[(f.name, f.fname) for f in fm.fontManager.ttflist if 'D2' in f.name]
```

```python
#(2) 폰트 적용하기. 
plt.rcParams["font.family"] = 'D2Coding'
plt.rcParams["font.size"] = 10
```

```python
#TIP. 그래프에서 -(minus sign) 표시
mpl.rcParams["axes.unicode_minus"]=False
```

```python
#TIP. 한글 폰트 적용 예
%matplotlib inline
import math
xs, ysin, ycos = [], [], []
for x in range(360):
        xs.append(x)
        ysin.append(math.sin(math.pi * x / 180))
        ycos.append(math.cos(math.pi * x / 180))
xs = np.arange(-6, 2*np.pi, 0.01)
ysin = np.sin(xs); ycos = np.cos(xs)
plt.plot(xs, ysin, label='Sine curve')
plt.plot(xs, ycos, label='Cosine curve')
plt.xlim([-6, 2*np.pi+0.1])
plt.ylim([-1.2, 1.2])
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.title('삼각함수')
plt.legend(loc=1)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.axvline(x=360, color='k', linewidth=0.5)
plt.show()
```

```python
#02. Line plot으로 데이터 시각화와 추가 옵션 파악하기.
#(1) y=x^2 그래프 그리기.
plt.plot([1,4,9,16])
plt.show()
```

```python
#(2) plot 제목 지정하기.
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("y=x^2 graph")
plt.show()
```

```python
#(3) 스타일 지정하기.
#①EX1
plt.plot([1,2,3,4],[1,4,9,16], c='r', marker='s', ls='--')
plt.title("y=x^2 graph with style(red, square, --)")
plt.show()
```

```python
#②EX2
plt.plot([1,2,3,4],[1,4,9,16], ls='-.', c='b', marker="^", ms=8, mec="r", mfc="r")
plt.title("y=x^2 graph with style")
plt.show()
```

```python
#(4) 수직(y)축, 수평(x)축 범위 지정하기.
plt.plot([1,2,3,4],[1,4,9,16], ls='-.', c='b', marker="^", ms=8, mec="r", mfc="r")
plt.title("y=x^2 graph with xlim, ylim")
plt.xlim(0,5)
plt.ylim(-2,20)
plt.show()
```

```python
#(5) 틱 설정하기. 
#    WHAT? 플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 하고 이 틱에 써진 숫자 혹은 글자를 틱 라벨(tick label)이라고 한다. 
x= np.arange(-np.pi,np.pi+0.01,0.01)
y= np.cos(x)
plt.plot(x,y)
plt.xlim(-3.3,3.3)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.ylim(-1.2,1.2)
plt.yticks([-1, 0, +1])
plt.title("cos graph set tick")
plt.show()
```

```python
#(6) 그리드 설정하기. 
x= np.arange(-np.pi,np.pi+0.01,0.01)
y= np.cos(x)
plt.plot(x,y)
plt.xlim(-3.3,3.3)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.ylim(-1.2,1.2)
plt.yticks([-1, 0, +1])
plt.grid()
plt.title("cos graph with grid")
plt.show()
```

```python
#(7) plot 여러 개 겹쳐 그리기. 
t = np.arange(0,5,0.2)
plt.plot(t,t)
plt.plot(t,0.5*t**2)
plt.plot(t,0.2*t**3)
plt.grid()
plt.title("Multiple graphs")
plt.show()
```

```python
#(8) 범례(Legend)
t = np.arange(0,5,0.2)
plt.plot(t,t,label="graph_1")
plt.plot(t,0.5*t**2,label="graph_2")
plt.plot(t,0.2*t**3,label="graph_3")
plt.grid()
plt.legend(loc=2)
plt.title("Multiple graphs with legend")
plt.show()
```

```python
#(9) 수직축, 수평축 라벨
t = np.arange(0,5,0.2)
plt.plot(t,t,label="graph_1")
plt.plot(t,0.5*t**2,label="graph_2")
plt.plot(t,0.2*t**3,label="graph_3")
plt.grid()
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Multiple graphs with x, y label")
plt.show()
```

```python
#03. Figure 객체
#(1) figsize를 설정해 플롯을 넓게 쓸려고 할 때 사용함.
np.random.seed(2021)
plt.figure(figsize=(10,2))
plt.title("figure size: (10,2)")
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

```python
#02. Axes 객체 
x1 = np.linspace(0.0,5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
plt.subplot(1,2,1)
plt.plot(x1, y1)
plt.grid()

x2 = np.linspace(0.0,2.0)
y2 = np.cos(2 * np.pi * x2)
plt.subplot(1,2,2)
plt.plot(x2, y2)
plt.grid()

plt.tight_layout()  
plt.show()
```

```python
np.random.seed(2021)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

np.random.seed(2021)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
#PLUS. for문과 axes 분할 응용하기. 
fig, axes = plt.subplots(2,2)

np.random.seed(2021)
for i in range(4)     :
    axes[i//2, i%2].plot(np.random.rand(5))
    axes[i//2, i%2].set_title(f"axes {i+1}")

plt.tight_layout()
plt.show()
```
