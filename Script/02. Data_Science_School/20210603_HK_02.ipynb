```python
#2021.06.03. THUR
#Hankyeong

##CH05-3. Seaborn 시각화
#00-1. 패키지 호출
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm 
import numpy as np 
import seaborn as sns
import matplotlib.image as img

#00-2. 시각화 옵션 설정하기. 
%matplotlib inline
mpl.rcParams["axes.unicode_minus"]=False
plt.rcParams["font.family"] = 'D2Coding'

#01. iris데이터셋을 countplot으로 변수의 범주별 개수 시각화하기.  
iris = sns.load_dataset('iris')
sns.countplot(x='species', data=iris)
plt.show()
```

```python
#02. iris 데이터셋을 pairplot으로 산점도, 히스토그램 파악하기. 
#(1) defualt 
sns.pairplot(iris)
plt.show()
```

```python
#(2) hue 파라미터로 카테고리 별로 구분하기. 
sns.pairplot(iris, hue="species")
plt.show()
```

```python
#03. iris 데이터셋의 연속형 변수 분포를 boxplot으로 시각화하기. 
iris.plot.box()
plt.show()
```

```python
#04. flight 데이터셋으로 heatmap 시각화하기. 
flights = sns.load_dataset('flights')
flights_passengers = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(8,6))
plt.title("연도, 월 별 승객수에 대한 Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()
```
