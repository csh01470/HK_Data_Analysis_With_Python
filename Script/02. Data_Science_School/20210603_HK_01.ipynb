```python
#2021.06.03. THUR
#Hankyeong

##CH05-2. Matplotlib의 다양한 Plot
#00-1. 패키지 호출
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm 
import numpy as np 
import seaborn as sns
import matplotlib.image as img

#00-2. 시각화 옵션 설정하기. 
%matplotlib inline
mpl.rcParams["axes.unicode_minus"]=False
plt.rcParams["font.family"] = 'D2Coding'

#01. 막대그래프(Bar chart)로 데이터 시각화와 추가 옵션 파악하기.
#(1) default 설정
y = [2, 3, 1]
x = np.arange(len(y))
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, ["가","나","다"])
plt.yticks(sorted(y))
plt.show()
```

```python
#(2) alpha 파라미터로 투명도 설정하기. 
y = [2, 3, 1]
x = np.arange(len(y))
plt.title("Bar Chart")
plt.bar(x, y, alpha=0.5)
plt.xticks(x, ["가","나","다"])
plt.yticks(sorted(y))
plt.show()
```

```python
#(3) barh 파라미터로 막대 눕히기. 
y = [2, 3, 1]
x = np.arange(len(y))
plt.title("Bar Chart")
plt.barh(x, y)
plt.yticks(x, ["가","나","다"])
plt.xticks(sorted(y))
plt.show()
```

```python
#02. 파이차트(pie chart)로 데이터 시각화와 추가 옵션 파악하기.
#(1) default
plt.title("Pie Chart")
labels = ['개구리', '돼지', '개', '소']
sizes = [15, 30, 45, 25]
plt.pie(sizes, labels=labels, autopct='%1.0f%%')
plt.axis('equal')
plt.show()
```

```python
#(2) color 파라미터로 색 지정하기. 
plt.title("Pie Chart")
labels = ['개구리', '돼지', '개', '소']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
sizes = [15, 30, 45, 25]
plt.pie(sizes, labels=labels, autopct='%1.0f%%', colors=colors)
plt.axis('equal')
plt.show()
```

```python
#(3) explode 파라미터로 부분 띄우기. 
plt.title("Pie Chart")
labels = ['개구리', '돼지', '개', '소']
sizes = [15, 30, 45, 25]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(sizes, labels=labels, autopct='%1.0f%%', colors=colors, explode= explode)
plt.axis('equal')
plt.show()
```

```python
#(4) shadow 파라미터로 그림자 지정하기. 
labels = ['개구리', '돼지', '개', '소']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)  
plt.axis('equal')
plt.show()
```

```python
#(5) startangle로 부분의 각도 조정하기. 
labels = ['개구리', '돼지', '개', '소']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)  
plt.axis('equal')
plt.show()
```

```python
#03. 히스토그램(histogram)으로 데이터 시각화하기. 
np.random.seed(2021)
x = np.random.randn(1000)
plt.title("Histogram")
plt.hist(x, bins=10)
plt.show()
```

```python
#04. 산점도(Scatter plot)으로 데이터 시각화하기.
np.random.seed(0)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.grid()
plt.show()
```

```python
#MEMO. 산점도는 두 변수 간의 상관관계를 파악할 때 사용함. 
```

```python
#05. 버블차트(Bubble Chart)로 데이터 시각화하기. 
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

```python
#MEMO. s 파라미터는 size를 의미함. 
```

```python
#06. 앤스컴콰르텟(Anscombe's Quartet)
#(1) 데이터셋 불러오기
ans = sns.load_dataset('anscombe')

#(2) dataset별 그룹핑하기. 
ans.groupby('dataset').describe()
```

```python
#MEMO. 그룹별로 y 변수의 평균, 표준편차가 동일함을 할 수 있음. 
```

```python
#(3) dataset별로 시각화하기. 
fig, axes = plt.subplots(2, 2, figsize=(10,8),
                         sharex=True, sharey=True)
for index, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.plot(ds.x, ds.y, 'o', markersize=10)
    ax.plot([3, 20], [4.5, 13], 'r-', lw=2) # y = 0.5 * x + 3
    ax.set_title(f'Dataset {roman}')

fig.suptitle("Anscombe's Quartet", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#RESULT. dataset별로 y변수의 평균, 표준편차는 동일하지만, 산포도가 각자 다름을 알 수 있음. 
```
