```python
#2021.06.04. FRI
#Hankyeong

#00. 패키지 호출하기.
import folium
import pandas as pd 

#03. folium 패키지를 이용해서, folium.Map() 함수 사용하기. 
#(1) default 
folium.Map(location=[37.5598688323642,126.96710965266])
```

```python
#(2) zoom_start 파라미터 사용하기.
folium.Map(location=[37.5598688323642,126.96710965266],zoom_start=20)
```

```python
#(3) tiles 파라미터 사용하기.
folium.Map(location=[37.5598688323642,126.96710965266],zoom_start=20,tiles="Stamen Toner")
```

```python
#02. folium 패키지를 이용해서, folium.Marker() 함수 사용하기. 
#(1) default
map = folium.Map(location=[37.5598688323642,126.96710965266],zoom_start=15,tiles="Stamen Toner")
folium.Marker(location=[37.5598688323642,126.96710965266]).add_to(map)
map
```

```python
#TIP. Marker() 함수를 사용하려면 기존의 Map() 함수를 사용한 결과를 변수로 지정해야함. 
```

```python
#(2) popup 파라미터 사용하기.
map = folium.Map(location=[37.5598688323642,126.96710965266],zoom_start=15,tiles="Stamen Toner")
folium.Marker(location=[37.5598688323642,126.96710965266],
              popup=folium.Popup("서울특별시 중구 청파로 463 (중림동)",max_width=1000)).add_to(map)
map
```

```python
#(3) tooltip 파라미터 사용하기. 
map = folium.Map(location=[37.5598688323642,126.96710965266],zoom_start=15,tiles="Stamen Toner")
folium.Marker(location=[37.5598688323642,126.96710965266],
              popup=folium.Popup("서울특별시 중구 청파로 463 (중림동)",max_width=1000),
              tooltip="한국경제신문").add_to(map)
map
```

```python
#PLUS. Icon() 함수로 표시기호를 변경할 수 있음.
map = folium.Map(location=[37.5598688323642,126.96710965266],zoom_start=15,tiles="Stamen Toner")
folium.Marker(location=[37.5598688323642,126.96710965266],
              popup=folium.Popup("서울특별시 중구 청파로 463 (중림동)",max_width=1000),
              tooltip="한국경제신문",
              icon=folium.Icon(color="red", icon="info-sign")).add_to(map)
map
```

```python
#03. folium 패키지를 이용해서, folium.Circle() 함수 사용하기. 
#(1) default
map = folium.Map(location=[37.5641201543296,126.998009728978],zoom_start=15,tiles="Stamen Toner")
folium.Circle(location=[37.5641201543296,126.998009728978],
              tooltip="중구청").add_to(map)
map
```

```python
#(2) color 파라미터 사용하기.
map = folium.Map(location=[37.5641201543296,126.998009728978],zoom_start=15,tiles="Stamen Toner")
folium.Circle(location=[37.5641201543296,126.998009728978],
              tooltip="중구청",
              color="crimson").add_to(map)
map
```

```python
#(3) fill 파라미터 사용하기. 
map = folium.Map(location=[37.5641201543296,126.998009728978],zoom_start=15,tiles="Stamen Toner")
folium.Circle(location=[37.5641201543296,126.998009728978],
              tooltip="중구청",
              color="crimson",
              fill=True).add_to(map)
map
```

```python
#PLUS, 01에서 추출했던 데이터프레임을 기반으로 지도 시각화하기. 
df = pd.read_csv("d:/python_project/hankyeong_dataanalysis/data/address.csv")
df
```

```python
map=folium.Map(location=[df["경도"].mean(),df["위도"].mean()], zoom_start=14.25, tiles="stamen toner")
for i in range(len(df)) :
    folium.Marker(location=[df["경도"][i],df["위도"][i]],
                  tooltip=df["빌딩명"][i],
                  popup=folium.Popup(df["도로명주소"][i],max_width=250)).add_to(map)
map
```
