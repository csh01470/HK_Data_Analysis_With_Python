```python
#2021.06.02. WED
#Hankyeong

##CH04-8. 시계열 자료 다루기
#00. 패키지 호출
import numpy as np 
import pandas as pd 

#01.   
#(1) 여러 데이터 타입 처리하기.
date_str = ["2018, 1, 1", "2018-1-4", "2018/1/5", "2018.1.6","20180107","010818"]
idx = pd.to_datetime(date_str)
idx
```

```python
#PLUS, 시계열 데이터를 인덱스로 설정하기. 
np.random.seed(2021)
s= pd.Series(np.random.random(6), index=idx)
s
```

```python
#(2) 시계열 데이터 정의하기. 
#①범위를 지정하기.
pd.date_range("20210601","20210630")
```

```python
#②periods 파라미터 사용하기. 
pd.date_range("20210601",periods=30)
```

```python
#(3) freq 파라미터로 조건 지정하기.
#①출석을 해야하는 날(Business Day) 출력하기. 
pd.date_range("20210601","20210630",freq="B")
```

```python
#②일요일 출력하기. 
pd.date_range("20210601","20210630",freq="W")
```

```python
#③토요일 출력하기. 
pd.date_range("20210601","20210630",freq="W-SAT")
```

```python
#02. resample 연산으로 시간 간격을 재조정, 그룹핑하기. 
ts = pd.Series(np.random.randn(365), index=pd.date_range(
    "2021-1-1", periods=365))
ts
```

```python
#(1) 주별 평균으로 값들 평균내 출력하기. 
ts.resample("W").mean()
```

```python
#(2) 각달의 마지막날 출력하기. 
ts.resample("M").first()
```
