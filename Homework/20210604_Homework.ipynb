```python
#2021.06.04. FRI
#Hankyeong

#00. 패키지 호출하기.
import folium
import pandas as pd 
import numpy as np 
import warnings
import requests
from urllib.parse import urlparse, quote

#00-1. warning message ignore
warnings.filterwarnings(action='ignore')

#01. 데이터프레임 불러오고 필요변수 추출하기.
#(1) 데이터셋 불러오기. 
park_raw = pd.read_csv("d:/python_project/hankyeong_dataanalysis/data/서울시_주요_공원현황(수정).csv")
park_raw.head(5)
```

```python
#(2) 필요변수만 추출하기. 
park = park_raw[["공원명","공원주소","면적","X좌표(WGS84)","Y좌표(WGS84)"]]
park
```

```python
#05. 면적 변수 단위 삭제 및 실수로 데이터 유형 변경하기. 
for i in range(len(park_rm_na)) :
    park_rm_na.iloc[i,2] = int(float(park_rm_na.iloc[i,2].replace("㎡","")))
```

```python
park_2 = park_rm_na
park_2
```

```python
park_2["category_2"] = pd.qcut(park_2["면적"],3,labels=["small","middle","big"])
park_2
```

```python
#06. 카카오 API를 사용해 위도, 경도값 파악하기.
#(1) kakao_api_key.txt 파일 부르기. 
with open("D:/Python_Project/Hankyeong_DataAnalysis/API_key/kakao_api_key.txt",mode="r") as key_file :
    kakao_api_key = key_file.read()
```

```python
#(2) for문으로 필요 데이터값 추출하기.
local_url = "https://dapi.kakao.com/v2/local/search/address.json"
x_list = []
y_list = []
error_i_list=[]
for i in park_2.index                                                          :
    url = f"{local_url}?query={quote(park_2['공원주소'][i])}"
    result = requests.get(url, headers={"Authorization" : f"KakaoAK {kakao_api_key}"}).json()
    try                                                                        :
        x = float(result['documents'][0]['x'])
        y = float(result['documents'][0]['y'])
        x_list.append(x)
        y_list.append(y)
    except Exception                                                            :
        error_i_list.append(i)
        print(f"ERROR ROW INDEX: {i}, PARK ADDRESS: {park_2['공원주소'][i]}")
```

```python
for i in error_i_list :
    park_2=park_2.drop(park_2.index[i])
park_2
```

```python
park_2["X좌표"]= x_list
park_2["Y좌표"]= y_list
park_2
```

```python
#06. 위도값,경도값을 KAKAO-API 기준으로 시각화하기. 
map = folium.Map(location=[park_2["Y좌표"].mean(),park_2["X좌표"].mean()],zoom_start=12,tiles="Stamen Toner")
for i in park_2.index        :
    if park_2["category_2"][i]=="big" :
        folium.Circle(location=[park_2["Y좌표"][i],park_2["X좌표"][i]],
        tooltip=F'{park_2["공원명"][i]}, {park_2["면적"][i]}㎡',
        fill=True,
        color="red",
        radius=300).add_to(map)
    if park_2["category_2"][i]=="middle" :
        folium.Circle(location=[park_2["Y좌표"][i],park_2["X좌표"][i]],
        tooltip=F'{park_2["공원명"][i]}, {park_2["면적"][i]}㎡',
        fill=True,
        color="green",
        radius=200).add_to(map)
    if park_2["category_2"][i]=="small" :
        folium.Circle(location=[park_2["Y좌표"][i],park_2["X좌표"][i]],
        tooltip=F'{park_2["공원명"][i]}, {park_2["면적"][i]}㎡',
        fill=True,
        color="blue",
        radius=100).add_to(map)
map
```
